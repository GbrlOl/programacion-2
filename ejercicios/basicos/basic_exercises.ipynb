# **[EIE409] Programación 2: Ejercicios Básicos**

### **Semana 1: Conceptos Básicos**
**Objetivos:**
- Familiarizarse con la sintaxis básica de Python.
- Trabajar con variables y tipos de datos.
- Realizar operaciones aritméticas y de cadena.

**Ejercicios:**
1. **Calculadora Básica:** Crea un programa que pida al usuario dos números y muestre el resultado de su suma, resta, multiplicación y división.
2. **Edad Futura:** Escribe un programa que pregunte tu nombre y edad, y muestre cuántos años tendrás dentro de 10 años.
3. **Área de un Rectángulo:** Desarrolla un programa que pida el ancho y alto de un rectángulo, y muestre su área.
4. **Saludo Personalizado:** Programa que pida el nombre del usuario y muestre un saludo personalizado.
5. **Manipulación de Cadenas:** Crea un programa que concatene dos cadenas y muestre el resultado en mayúsculas y minúsculas.

---

### **Semana 2: Estructuras de Control**
**Objetivos:**
- Comprender y utilizar condicionales (`if`, `elif`, `else`).
- Trabajar con bucles (`for`, `while`).
- Aplicar estructuras de control en problemas cotidianos.

**Ejercicios:**
1. **Juego de Adivinanza:** Crea un juego donde el usuario adivine un número secreto entre 1 y 10.
2. **Suma de Números:** Programa que pida números al usuario hasta que ingrese un número negativo, y muestre la suma de los números positivos.
3. **Lista de Nombres:** Escribe un programa que pida nombres al usuario y los almacene en una lista, mostrando la lista ordenada al final.
4. **To-Do List:** Desarrolla una aplicación de lista de tareas que permita agregar y eliminar tareas.
5. **Impresión de Números:** Programa que muestre los números del 1 al 100 usando un bucle `for`.

---

### **Semana 3: Funciones y Estructuras de Datos**
**Objetivos:**
- Crear y utilizar funciones para organizar el código.
- Trabajar con listas y tuplas.
- Realizar operaciones básicas con estas estructuras de datos.

**Ejercicios:**
1. **Suma de Números con Función:** Crea una función que reciba una lista de números y retorne su suma.
2. **Calculadora con Funciones:** Escribe funciones separadas para suma, resta, multiplicación y división, y crea un menú para que el usuario seleccione la operación.
3. **Verificador de Par/Impar:** Programa que use una función para determinar si un número es par o impar.
4. **Ordenación de Listas:** Crea un programa que pida una lista de números y muestre la lista ordenada de menor a mayor.
5. **Manipulación de Tuplas:** Escribe un programa que trabaje con tuplas, demostrando su inmutabilidad y operaciones básicas.

---

### **Semana 4: Archivos y Excepciones**
**Objetivos:**
- Leer y escribir archivos de texto.
- Manejar excepciones para controlar errores en el programa.
- Trabajar con diferentes formatos de archivo como CSV y JSON.

**Ejercicios:**
1. **Lector de Archivos:** Programa que lea un archivo de texto y muestre su contenido.
2. **Escritor de Archivos:** Crea un programa que escriba en un archivo las frases ingresadas por el usuario.
3. **Gestión de Listas en Archivos:** Escribe un programa que permita agregar, eliminar y mostrar elementos de una lista almacenada en un archivo.
4. **Excepciones Básicas:** Programa que maneje excepciones al dividir dos números, evitando que el programa se cierre si hay un error.
5. **Trabajo con CSV:** Crea un programa que lea un archivo CSV y muestre su contenido en formato tabular.
6. **Trabajo con JSON:** Escribe un programa que guarde y lea datos en formato JSON, mostrando la información estructurada.

---

### **Proyecto Final**
**Objetivo:**
- Aplicar todos los conceptos aprendidos en un proyecto integral.

**Ejercicio:**
1. **Juego de Aventuras:** Crea un juego de texto donde el usuario tome decisiones que afecten el resultado final. El juego debe incluir:
   - Menú de opciones.
   - Condicionales para diferentes caminos en la historia.
   - Funciones para cada parte del juego.
   - Lectura y escritura de archivos para guardar y cargar partidas.
   - Manejo de excepciones para errores inesperados.

---

Este plan de ejercicios está diseñado para ser progresivo, permitiendo al principiante construir una base sólida en Python. Cada semana introduce nuevos conceptos y desafíos, culminando en un proyecto que integra todo lo aprendido.
